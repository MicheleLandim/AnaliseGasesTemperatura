```python
dtf = pd.read_csv('./Banco_dados.csv', low_memory=False,delimiter=';',  encoding='iso-8859-1', decimal=',')
print("Banco de dados com as informações de cada mesorregião.")
dtf.head()
```

```python
import matplotlib.pyplot as plt
legendas = 'Nordeste Sudeste Sul Norte Centro-Oeste'.split()
print (dtf.head())
```

```python
plt.axis('equal')
plt.pie(dtf['LOCAL'].value_counts(), autopct="%.2f",
        labels=legendas, explode = [0.1,0,0,0,0])
plt.suptitle("Porcentagem de mesorregiões por Região")

plt.show()
```

```python
dtProdEner = pd.read_csv('./ProducaoEnergia.csv', low_memory=False,delimiter=';',  encoding='iso-8859-1', decimal=',')
print("Banco de dados com informações de Temperatura e quantidade de CH4, CO2 e N2O emitido por ano.")
dtProdEner.head()
```

```python
dt_f = dtf.groupby(['tipo','Ano']).apply(lambda x: x['Qtd'].sum())
dtf_temperatura = pd.read_csv('./ProducaoEnergia.csv', low_memory=False,delimiter=';', encoding='UTF-8', decimal=',')
dfu = dt_f.unstack(level=0)
dfu = dfu.drop('CH4', axis=1)
dfu = dfu.drop('N2O', axis=1)

dtf_AnoTemp = dtf_temperatura["Ano"]
dtf_MaxTemp = dtf_temperatura["Temp_Max"].mul(100000)
dtest = pd.DataFrame(dtf_temperatura["Temp_Max"])

d = {'Ano1':dtf_AnoTemp,'Temp':dtf_MaxTemp}
d1 = {'Ano1':dfu.index, 'CO2':dfu['CO2']}

dtf_AnoMaxTemp = pd.DataFrame(d)
dfu = pd.DataFrame(d1)
dtf_novo = pd.merge(dtf_AnoMaxTemp, dfu)
dtf_novo = dtf_novo.set_index('Ano1')

dtf_AnoMaxTemp = dtf_AnoMaxTemp.set_index('Ano1')
dfu = dfu.set_index('Ano1')

dfu.plot()
plt.legend(loc='best')
plt.title("Emissão de CO2", weight='bold')
plt.xlabel('Ano')
plt.ylabel('Quantidade em Gigagrama.')

dtf_AnoMaxTemp.plot()
plt.legend(loc='best')
plt.title("Temperatura Máxima vezes 100000", weight='bold')
plt.xlabel('Ano')
plt.ylabel('Celsius')

dtf_novo.plot()
plt.legend(loc='best')
plt.title("Temperatura vezes 100000 X CO2", weight='bold')
plt.xlabel('Ano')
plt.ylabel('qtd')
plt.show()
```

```python
import matplotlib.pyplot as plt
 
dtf = pd.read_csv('./ProducaoEnergia.csv', low_memory=False,delimiter=';', encoding='UTF-8', decimal=',')
dtf.head()
dt_f = dtf.groupby(['tipo','Ano']).apply(lambda x: x['Qtd'].sum())
dfu = dt_f.unstack(level=0)
'print(dfu)'
dfu.plot()
plt.legend(loc='best')
plt.title("Quantidade de CO2, N2O e CH4 por ano no setor de produção de energia - Resumo.", weight='bold')
plt.xlabel('Ano')
plt.ylabel('Quantidade')
plt.show()

plt.title("Quantidade de CH4 emitido por ano na produção de energia.", weight='bold')
plt.plot(dfu["CH4"], label = 'CH4', color='blue')
plt.xlabel('Ano')
plt.ylabel('Quantidade')
plt.show()

plt.title("Quantidade de CO2 emitido por ano na produção de energia.", weight='bold')
plt.plot(dfu["CO2"], label = 'CO2', color='orange')
plt.xlabel('Ano')
plt.ylabel('Quantidade')
plt.show()

plt.title("Quantidade de N2O emitido por ano na produção de energia.", weight='bold')
plt.plot(dfu["N2O"], label = 'N2O', color='green')
plt.xlabel('Ano')
plt.ylabel('Quantidade')
plt.show()


```
